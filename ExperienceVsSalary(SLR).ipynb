```python
# step -1: Data preprocessing 

# 1.1 import libraries 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# 1.2 import dataset
dataset  = pd.read_csv('D:\Career\courses\Machine Learning A to Z Course\Dataset\Machine Learning A-Z (Codes and Datasets)\Part 2 - Regression\Section 4 - Simple Linear Regression\Python\Salary_Data.csv')
dataset
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, 1].values

# 1.3 spliting dataset into the Traing data set and  Testing data set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size  = 1/3, random_state = 0)

# 1.3 feature scaling
#from sklearn.preprocessing import StandardScaler

# sc_X = StandardScaler()
# X_train = sc_X.fit_transform(X_train)
# X_test = sc_X.transform(X_test)
# X_train
```

```python
# step -2: fitting simple linear regression to the training set

# 2.1 Train the model
from sklearn.linear_model import LinearRegression
regressor  = LinearRegression()
regressor.fit(X_train, Y_train)
```

```python
# predicting the Test set result

# 2.2 Predic the result 
y_pred = regressor.predict(X_test)
y_pred
```

```python
Y_test
```

```python
# setp-3: visualizing the traing set results

# 3.1 Training set
plt.scatter(X_train, Y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title("Salary vs Experience(Training set)")
plt.xlabel("Years of Experinces")
plt.ylabel("Salary")
plt.show()
```

```python
# visualizing the Test set results

# 3.2 Testing set
plt.scatter(X_test, Y_test, color = 'red')
plt.plot(X_test, y_pred , color = 'blue')
plt.title("Salary vs Experience(Testing set)")
plt.xlabel("Years of Experinces")
plt.ylabel("Salary")
plt.show()
```
